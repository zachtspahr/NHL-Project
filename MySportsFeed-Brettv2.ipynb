```python
## Building a loop to go through each season,each team, each game
## What is missing is the full list of team IDS, and including a Try and Except.
## we can use this lopp to build the dataframe of the stats we need.
##This is a limited test over three seasons

#Defining Lists
ListSeason=[]
ListTeam=[]
ListTeamName=[]
ListblockedShots=[]
ListgoalsAgainst=[]
ListpenaltyKillPercent=[]
ListPenaltyMinutes=[]
ListpenaltyKillgoals=[]
Listpenalties = []
ListWin=[]
ListfaceoffPercent=[]
ListfaceoffWins=[]
ListfaceoffLosses=[]
Listshots = []
Listhits = []

#Defining seasons & teams
TargetSeasons=["2015-2016-regular","2016-2017-regular"]
teams=['SJS','STL','CBJ','NSH','TOR','WPJ','MTL','DAL','COL','FLO','ARI','CHI','MIN','PHI','ANA','EDM','VAN','NYR','BUF','DET','NJD','LAK','OTT','TBL','CGY','BOS','WSH','NYI','PIT','CAR']
# 'VGK',2017-2018-regular,2018-2019-regular
#Creating loops to fill the lists
try:
    for season in TargetSeasons:
        for team in teams:
            for game in range (0,83):
                try:
                    TeamGameLogs = msf.msf_get_data(league='nhl',season=season,feed='team_gamelogs',team=team, format='json') 
                    teamname = TeamGameLogs["teamgamelogs"]["gamelogs"][game]["team"]["Abbreviation"]
                    BlockedShots=TeamGameLogs["teamgamelogs"]["gamelogs"][game]["stats"]["BlockedShots"]["#text"]
                    win=TeamGameLogs["teamgamelogs"]["gamelogs"][game]["stats"]["Wins"]["#text"]
                    goalsAgainst=TeamGameLogs["teamgamelogs"]["gamelogs"][game]["stats"]["GoalsAgainst"]["#text"]
                    PenaltyKillPercent=TeamGameLogs["teamgamelogs"]["gamelogs"][game]["stats"]["PenaltyKillPercent"]["#text"]
                    penalties=TeamGameLogs["teamgamelogs"]["gamelogs"][game]["stats"]["Penalties"]["#text"]
                    penaltyminutes=TeamGameLogs["teamgamelogs"]["gamelogs"][game]["stats"]["PenaltyMinutes"]["#text"]
                    penaltykillgoals=TeamGameLogs["teamgamelogs"]["gamelogs"][game]["stats"]["PenaltyKillGoalsAllowed"]["#text"]
                    FaceoffPerc=TeamGameLogs["teamgamelogs"]["gamelogs"][game]["stats"]["FaceoffPercent"]["#text"]
                    Faceoffwin=TeamGameLogs["teamgamelogs"]["gamelogs"][game]["stats"]["FaceoffWins"]["#text"]
                    Faceoffloss=TeamGameLogs["teamgamelogs"]["gamelogs"][game]["stats"]["FaceoffLosses"]["#text"]
                    shots=TeamGameLogs["teamgamelogs"]["gamelogs"][game]["stats"]["Shots"]["#text"]
                    hits=TeamGameLogs["teamgamelogs"]["gamelogs"][game]["stats"]["Hits"]["#text"]
                    
                except:
                    continue

                ## Adding values to list
                ListSeason.append(season)
                ListTeam.append(team)
                ListTeamName.append(teamname)
                ListblockedShots.append(BlockedShots)
                ListWin.append(win)
                ListgoalsAgainst.append(goalsAgainst)
                ListpenaltyKillPercent.append(PenaltyKillPercent)
                Listpenalties.append(penalties)                
                ListPenaltyMinutes.append(penaltyminutes)
                ListpenaltyKillgoals.append(penaltykillgoals)
                ListfaceoffPercent.append(FaceoffPerc)
                ListfaceoffWins.append(Faceoffwin)
                ListfaceoffLosses.append(Faceoffloss)
                Listshots.append(shots)
                Listhits.append(hits)
except:
    print("Error Found")
    
    
print ("Completed")
```

```python
#zipping the lists and creating a dataframe

ZipList=list(zip(ListSeason, ListTeam, ListTeamName, ListblockedShots,ListgoalsAgainst,ListpenaltyKillPercent,Listpenalties,ListPenaltyMinutes,ListpenaltyKillgoals,ListWin,ListfaceoffPercent,ListfaceoffWins,ListfaceoffLosses,Listshots,Listhits))

Gamesdf = pd.DataFrame(ZipList, columns = ["Season",
                                           "Team",
                                           "Team Name",
                                           "Blocked Shots",
                                           "Goals Against",
                                           "Penalty Kill Percent",
                                           "Penalties",
                                           "Penalty Minutes",
                                           "Penalty Kill Goals Allowed",
                                           "Win",
                                           "Faceoff Percent",
                                           "Faceoff Wins",
                                           "Faceoff Losses",
                                           "Shots",
                                           "Hits"])

Gamesdf["Win"]=Gamesdf["Win"].astype('int32')
Gamesdf["Blocked Shots"]=Gamesdf["Blocked Shots"].astype('int32')
Gamesdf["Goals Against"]=Gamesdf["Goals Against"].astype('int32')
Gamesdf["Penalty Kill Percent"]=Gamesdf["Penalty Kill Percent"].astype('float')
Gamesdf["Penalties"]=Gamesdf["Penalties"].astype('int32')
Gamesdf["Penalty Minutes"]=Gamesdf["Penalty Minutes"].astype('int32')
Gamesdf["Penalty Kill Goals Allowed"]=Gamesdf["Penalty Kill Goals Allowed"].astype('int32')
Gamesdf["Faceoff Percent"]=Gamesdf["Faceoff Percent"].astype('float')
Gamesdf["Faceoff Wins"]=Gamesdf["Faceoff Wins"].astype('int32')
Gamesdf["Faceoff Losses"]=Gamesdf["Faceoff Losses"].astype('int32')
Gamesdf["Shots"]=Gamesdf["Shots"].astype('int32')
Gamesdf["Hits"]=Gamesdf["Hits"].astype('int32')
export_csv = Gamesdf.to_csv ('2015-2017_gamelog.csv', index = None, header=True)
Gamesdf.head(5)
```

```python
#Grouping by

teamstats=Gamesdf.groupby("Team Name").agg({
    'Blocked Shots':sum,
    'Goals Against':sum,
    'Penalty Kill Percent':"mean",
    'Penalties':sum,   
    'Penalty Minutes':sum,
    'Penalty Kill Goals Allowed':sum,
    'Faceoff Percent':"mean",
    'Faceoff Wins':sum,
    'Faceoff Losses':sum,
    'Shots':sum,   
    'Hits':sum,
    })
teamstats["PK Goals Allows/Min"] = teamstats["Penalty Kill Goals Allowed"] / teamstats["Penalty Minutes"]
teamstats["FO %"] = teamstats["Faceoff Wins"] / (teamstats["Faceoff Wins"] + teamstats["Faceoff Losses"])*100
teamstats=teamstats.drop(['Penalty Kill Percent','Penalty Minutes','Penalty Kill Goals Allowed','Faceoff Percent','Faceoff Wins','Faceoff Losses'], axis=1)
export_csv = teamstats.to_csv ('2015-2017_teamstats.csv', header=True)
teamstats
```

```python
#Grouping by win/loss to compare stats

WinLoss=Gamesdf.groupby("Win").mean()
WinLoss["PK Goals Allows/Min"] = WinLoss["Penalty Kill Goals Allowed"] / WinLoss["Penalty Minutes"]
WinLoss["FO %"] = WinLoss["Faceoff Wins"] / (WinLoss["Faceoff Wins"] + WinLoss["Faceoff Losses"])*100
WinLoss=WinLoss.drop(['Penalty Kill Percent','Penalty Minutes','Penalty Kill Goals Allowed','Faceoff Percent','Faceoff Wins','Faceoff Losses'], axis=1)
export_csv = WinLoss.to_csv ('2015-2017_WinLoss.csv', header=True)
WinLoss
```
